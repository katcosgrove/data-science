# 3.) Given all the weather data here, find the average precipitation per month, and the median precipitation.

```python
weather['Date'] = pd.to_datetime(weather['Date'], errors='coerce')
```

```python
weather['Month'] = weather['Date'].dt.month
```

# 5.) When a bike station is modified, is it more likely that it’ll lose bikes or gain bikes? How do you know?

```python
station['modification_date'] = station['modification_date'].apply(lambda x: 0 if x is 'NaN' else 1)
```

```python
modified = pd.DataFrame(station[['station_id','modification_date', 'install_dockcount', 'current_dockcount']])
modified_filter = modified.query('modification_date == 1')
```

```python
greater = modified_filter.query('current_dockcount > install_dockcount').count()
greater
```

```python
less = modified_filter.query('current_dockcount < install_dockcount').count()
less
```

## Answer: It is more likely to lose bikes.
